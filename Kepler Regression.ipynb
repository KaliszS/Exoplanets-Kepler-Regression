# Etap 3

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold 
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
df = df.drop(['ID','Eccentricity'], axis = 1)
```

```python
df[df['Exoplanet Archive Disposition'] == 'CONFIRMED'].count().max()
```

```python
df[df['Exoplanet Archive Disposition'] == 'FALSE POSITIVE'].count().max()
```

```python
df[df['Exoplanet Archive Disposition'] == 'CANDIDATE'].count().max() / (2303+4841)
```

```python
df['Exoplanet Archive Disposition'].unique()
disposition = {'CONFIRMED': 1, 'CANDIDATE': 0.33874580067189247,'FALSE POSITIVE': 0}
df['Exoplanet Archive Disposition'] = df['Exoplanet Archive Disposition'].map(disposition)
```

```python
df['Planetary Fit Type'].unique()
fit_type = {'LS+MCMC': 3, 'MCMC': 2,'LS': 1, 'none': 0}
df['Planetary Fit Type'] = df['Planetary Fit Type'].map(fit_type)
```

```python
newX = df.drop('Stellar Effective Temperature', axis=1)
newY = df['Stellar Effective Temperature']
```

```python
X_train,X_test,y_train,y_test=train_test_split(newX,newY,test_size=0.3,random_state=3)
print(len(X_test), len(y_test))
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
rr = Ridge(alpha=0.01, max_iter = 20000)
rr.fit(X_train, y_train)
```

```python
lass = Lasso(alpha=0.01, max_iter = 20000)
lass.fit(X_train, y_train)
```

```python
train_score=lr.score(X_train, y_train)
test_score=lr.score(X_test, y_test)

Ridge_train_score = rr.score(X_train,y_train)
Ridge_test_score = rr.score(X_test, y_test)

Lasso_train_score = lass.score(X_train,y_train)
Lasso_test_score= lass.score(X_test, y_test)
```

```python
print("linear regression train score:", train_score)
print("linear regression test score:", test_score)
print("ridge regression train score low alpha:", Ridge_train_score)
print("ridge regression test score low alpha:", Ridge_test_score)
print("Lasso regression train score low alpha:", Lasso_train_score)
print("Lasso regression test score low alpha:", Lasso_test_score)
```

```python
print("Walidacja Krzyżowa dla regresji liniowej :", cross_val_score(lr, newX, newY, cv=10)[1:])
print("Walidacja Krzyżowa dla regresji z reg. grzbietową :", cross_val_score(rr, newX, newY, cv=10)[1:])
print("Walidacja Krzyżowa dla regresji z reg. lasso :", cross_val_score(lass, newX, newY, cv=10)[1:])
```

```python
y_pred = lass.predict(X_test)
plt.scatter(y_pred, range(len(y_pred)), color="red")
plt.scatter(y_test, range(len(y_test)), color='blue')
plt.show()
```

```python
y_pred = rr.predict(X_test)
plt.scatter(y_pred, range(len(y_pred)), color="red")
plt.scatter(y_test, range(len(y_test)), color='blue')
plt.show()
```

```python
plt.plot(rr.coef_,alpha=0.7,linestyle='none',marker='*',markersize=5,color='red',label=r'Ridge; $\alpha = 0.01$',zorder=7) # zorder for ordering the markers
plt.plot(lr.coef_,alpha=0.4,linestyle='none',marker='o',markersize=7,color='green',label='Linear Regression')
plt.xlabel('Coefficient Index',fontsize=16)
plt.ylabel('Coefficient Magnitude',fontsize=16)
plt.legend(fontsize=13,loc=4)
plt.show()
```
